```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import os
import shutil
import matplotlib.pyplot as plt

# --- 1. Data Restructuring (Same as your code) ---
# This step is correct and necessary to get the 6 classes at the top level.
src_dir = "/kaggle/input/studentengagement/Student-engagement-dataset"
data_dir = "/kaggle/working/engagement_flattened"

if os.path.exists(data_dir):
    shutil.rmtree(data_dir) # Clean up previous runs
os.makedirs(data_dir, exist_ok=True)

# Loop through 'Engaged' & 'Not engaged' subfolders to flatten the structure
for category in os.listdir(src_dir):
    category_path = os.path.join(src_dir, category)
    if os.path.isdir(category_path):
        for subcat in os.listdir(category_path):
            subcat_path = os.path.join(category_path, subcat)
            if os.path.isdir(subcat_path):
                target_path = os.path.join(data_dir, subcat)
                # Instead of making a new dir for each image, copy the whole folder
                shutil.copytree(subcat_path, target_path)

print("Data restructured. Available classes:")
print(os.listdir(data_dir))

# --- 2. Data Loading and Preprocessing with TensorFlow/Keras ---
# Define parameters
IMG_SIZE = (224, 224) # Using 224x224 to match your PyTorch transforms
BATCH_SIZE = 32
img_shape = (IMG_SIZE[0], IMG_SIZE[1], 3)

# Load datasets from directory
# This function automatically creates train/validation splits and infers class labels.
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

class_names = train_ds.class_names
print("\nClasses found by Keras:", class_names)
num_classes = len(class_names)

# --- 3. Data Augmentation and Performance Tuning ---
# Create a data augmentation layer
data_augmentation = keras.Sequential([
    keras.layers.RandomFlip("horizontal"),
    keras.layers.RandomRotation(0.1), 
    keras.layers.RandomZoom(height_factor=0.2, width_factor=0.2),
    keras.layers.RandomTranslation(height_factor=0.1, width_factor=0.1), 
    keras.layers.RandomBrightness(factor=0.2),
    keras.layers.RandomContrast(factor=0.2), 
])

# Apply augmentation only to the training set
train_ds = train_ds.map(lambda x, y: (data_augmentation(x, training=True), y))

# Rescale pixel values from [0, 255] to [0, 1] for both datasets
rescaling = keras.layers.Rescaling(1./255)
train_ds = train_ds.map(lambda x, y: (rescaling(x), y))
val_ds = val_ds.map(lambda x, y: (rescaling(x), y))

# Configure dataset for performance
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)


# --- 4. Model Definition (Your specified architecture) ---
model = Sequential([
    # Input Layer
    keras.layers.Input(shape=img_shape),

    # Convolutional Blocks
    Conv2D(32, (1, 1), activation='relu'),
    MaxPooling2D(3, 3),
    Conv2D(64, (1, 1), activation='relu'),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(3, 3),
    Conv2D(128, (1, 1), activation='relu'),
    Conv2D(128, (5, 5), activation='relu'),
    MaxPooling2D(3, 3),
    Conv2D(256, (1, 1), activation='relu'),
    Conv2D(256, (5, 5), activation='relu'),

    # Classifier Head
    Flatten(),
    Dense(512, activation='relu'),
    Dense(num_classes, activation='softmax') # Use num_classes for flexibility
])

model.summary()

# --- 5. Model Compilation ---
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="sparse_categorical_crossentropy", # Use this loss for integer labels
    metrics=['accuracy']
)

# --- 6. Callbacks ---
# Save the best model based on validation accuracy
checkpoint = ModelCheckpoint(
    "Student_Engagement_Model.h5",
    monitor='val_accuracy',
    verbose=1,
    save_best_only=True,
    mode='max'
)

# Stop training early if validation loss doesn't improve
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10, # Number of epochs with no improvement
    verbose=1,
    restore_best_weights=True
)


# --- 7. Model Training ---
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=100, # Set a high number, EarlyStopping will handle the rest
    callbacks=[checkpoint, early_stopping]
)


# --- 8. Visualize Results ---
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(len(acc))

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
